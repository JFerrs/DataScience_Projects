```python
import pandas as pd
melb_df = pd.read_csv('melb_data.csv')
print(melb_df)
```

```python
melb_df.describe()
```

```python
#limpiamos, quitando las filas que tengan Na
melb_df = melb_df.dropna(axis=0)
melb_df
```

```python
#Vamos a definir las variables a predecir y las que nos sirven para ello
y= melb_df['Price'] #Variable a predecir
X= melb_df[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']] #Variables numericas que nos sirven por su importancia
```

```python
#Utilizamos Regresor porque queremos predecir una variable numérica. Decision Tree va analizsando cada una de las caracteristicas que le hemos indicado  
from sklearn.tree import DecisionTreeRegressor
melbourne_model = DecisionTreeRegressor() #Establecemos el modelo vacio
melbourne_model.fit(X, y) #Entrenamos el modelo, pasandole las caracteristicas que hemops establecido anteriomente
```

```python
#Ahora que hemos entrenado el modelo, vamos a ver algunas predicciones
print('Predicciones para las primeras 5 casas: ')
print(X.head())
print('El precio real es: ')
print(y.head())
print('Las predicciones son ')
print(melbourne_model.predict(X.head()))
```

```python
# Vemos que los precios de la predicion y los reales coinciden en las 5 primeras casas, por lo que el modelo parece que funciona bien
# Vamos a comprobar ahora con todas, para observar el error
# Utilizamos el MAE (mean absolute error), calculamos el error que está cometiendo para cada casa, obteniendo la media absoluta.
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error (y, predicted_home_prices)
```

```python
# Vemos que por ahora el error es muy bajo, ya que comparado con el valor de las casas, este es muy bajo (1ooo frente a 1000000). Es un buen modelo.
# Vamos a validarlo, dividiendo los datos que tenemos en datos para entrenar y datos para validar
from sklearn.model_selection import train_test_split 
train_X, val_X, train_y, val_y = train_test_split(X, y)
melbourne_model = DecisionTreeRegressor() # Creamos entonces otro modelo, entrenandolo solamente con el train.
melbourne_model.fit(train_X, train_y)
val_predictions= melbourne_model.predict(val_X) # Una vez entrenado, vamos a compara los precios del train de X con los reales de y, para ver el MAE.
print(mean_absolute_error(val_y, val_predictions))
```

```python
# Vemos que este modelo en realidad no es tan bueno, ya que el error ha subido bastante. Intentemos arreglarlo cambiando parametros del entrenamiento
# Creamos una funcion para el entrenamiento
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes)
    model.fit(train_X, train_y)
    preds_val =model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# Creamos un buble for para variar la cantidad de hojas que queremos comprobar
for max_leaf_nodes in [5, 50, 500, 1000, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print('Max leaf nodes: %d \t\t Mean Absolute Error: %d' %(max_leaf_nodes, my_mae))
```

```python
# Vamos a compararlo con otro algoritmo, en este caso un random forest
from sklearn.ensemble import RandomForestRegressor 
forest_model = RandomForestRegressor() # Creamos entonces otro modelo, entrenandolo solamente con el train.
forest_model.fit(train_X, train_y)
melb_preds= forest_model.predict(val_X) # Una vez entrenado, vamos a compara los precios del train de X con los reales de y, para ver el MAE.
print(mean_absolute_error(val_y, melb_preds))
```

```python
# Vemos que solo cambiando el modelo a usar, sin variar parámetros internos el error es menor. 
# A veces es necesario aplicar varios modelos diferentes, porque pueden ser mejores unos que otros según los datos. 
```
